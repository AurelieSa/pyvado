You just implemented design wth pyvado but you don't know if your design work? Okay let's do a simulation with pyvado!

```python
import pyvado
```

```python
pv = pyvado.Pyvado("./my_vivado_project/my_vivado_project.xpr")
pv.project.open()
```

First we need to add our test bench and set up as simulation toplevel.

```python
pv.files.add_simulation_file("./design_files/toplevel_test_tesbench.vhd")
pv.simulator.set_toplevel("toplevel_test_tb")
```

Now we can open the vivado simulator.

```python
pv.simulator.open()
```

Now run simulation!

```python
pv.simulator.run("20ns")
```

Ok...but I want to record acitivity!

Okay Okay don't worry, let's open a activity file.

```python
pv.simulator.restart()
pv.simulator.log_saif("./outputs/activity.saif")
```

And now we can run again our simulation.

```python
pv.simulator.run("20ns")
```

Perfect! You just run simulation with Pyvado!
