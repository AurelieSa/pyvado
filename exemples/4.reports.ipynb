You just synthesis your first circuit. Now we will see how we can get some reports. 

First we need to create a pyvado session and open project.

```python
import pyvado
```

```python
pv = pyvado.Pyvado("./my_vivado_project/my_vivado_project.xpr")
pv.project.open()
pv.flow.reset_run()
```

First we need to check if we can open some runs. To do that, we will use `get_runs()` method.

```python
pv.report.get_runs()
```

Oh...no runs to open...ok let's run synthesis before report.

```python
pv.files.add_file(file_path=[
  "./design_files/toplevel_test.vhd",
  "./design_files/const.xdc" # don't worry we manage this
])
pv.flow.set_toplevel("toplevel_test")
pv.flow.synthesis()
```

Oki, let see if we got run to open.

```python
pv.report.get_runs()
```

Perfect! We will get reports of synthesis `synth_1`.

```python
pv.report.open(
  run_name="synth_1"
)
```

Report is open. We can get utilization and power report (don't worry I'll code other report soon).

```python
pv.report.utilization(
  output_path="./outputs/example_uti.txt",
  hierarchical=True
)
```

```python
pv.report.power(
  output_path="./outputs/example_power.txt"
)
```

Perfect! You just generate your first reports!

But you want to use a saif file for you power report? Let's do that!

```python
pv.report.set_activity("./outputs/activity.saif")
```

Now we can run again the power report.

```python
pv.report.power("./outputs/example_power_with_activity.saif")
```
