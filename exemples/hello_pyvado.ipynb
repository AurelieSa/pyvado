Welcom to Pyvado Hellow World Script

In this script, we will see basic command of pyvado.

First, import pyvado

```python
from pyvado import Pyvado
```

Now we will open our pyvado session

```python
pv = Pyvado()
```

Perfect! We will run our first TCL command. A simple puts command.

```python
pv.run_command(
  cmd = "puts \"Hello Pyvado!\"",
  blocking=False
)
```

Now we will read the output of vivado consol

```python
pv.read_output()
```

Congrats! You run your first command with Pyvado!

Now we will run two commands. To do that we can call run_command two times or we can send a list of commands

```python
cmd = [
  "puts \"Command 1\"",
  "puts \"Command 2\""
]
pv.run_command(
  cmd = cmd,
  blocking=False
)
```

```python
print(f"result of command 1 {pv.read_output()}")
print(f"result of command 2 {pv.read_output()}")
```
