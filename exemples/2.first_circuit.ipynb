In this examples, you will synthesis and deploy your first circuit on your FPGA. For this tutorial, we assume you create a vivado project and hardware design files to synthesis.

First, import pyvado and create instance.

````python
import pyvado
````

This time, we will open pyvado with a specific vivado project. Give to pyvado the vivado project path. Replaces values for your project and files. I personnaly create a simple project for Nexys A7-100T FPGA.

````python
pv = pyvado.Pyvado(project_path="my_vivado_project/my_vivado_project.xpr") # change the path to your project (don't worry pyvado will get the absolute path)
````

Now we need to open our project.

````python
pv.project.open()
````

First step, we need to add our files. It is importante to notice in this exemple I use project for Nexys A7-100T, even if I share with you design files you can replaces this tutorials with your own files.

To add files, we'll use the file manager.

````python
pv.files.add_file(
  file_path="./design_files/toplevel_test.vhd",
  used_in_synth=True,
  used_in_sim=True,
  import_file=False
)
````

We can check if our file is in the project. To do that run ```get_files()``` command

````python
pv.files.get_files()
````

We also need to set constrainst files. You can use `add_file()` or a method wrapper `add_constraint_file()`

````python
pv.files.add_constraint_file(
  file_path="./design_files/const.xdc"
)
````

Now we will setup toplevel to run synthesis. To do thaht we will use flow manager

````python
pv.flow.set_toplevel(
  module_name="toplevel_test"
)
````

Now we can run synthesis! First we need to reset synthesis

````python
pv.flow.reset_run()
````

First step, run synthesis.

````python
pv.flow.synthesis()
````

Synthesis done! Now let's go for implementation.

````python
pv.flow.implementation()
````

Perfect! Last step, generate bitstream!

````python
pv.flow.bitstream()
````

Now we will deploy our circuit to the board. To do that we will use the hardware manager.

````python
pv.hardware.open_hardware()
````

````python
pv.hardware.connect_server()
````

````python
pv.hardware.open_target()
````

Target is open, we can set up the bitstream

Ok but I don't remember/know the .bit file!

Don't worry! **If a project is open** pyvado will search this file by their own.

````python
pv.hardware.set_bitstream()
pv.hardware.program_device()
````

And that its! You send your first bitstream to your FPGA with Pyvado...yeah..but...so many commands!

Don't worry, we will do it with two commands!

The first one is `run_all()` that reset runs + run synthesis + implementation + bitstream

````python
pv.flow.run_all()
````

Second : `deploy()` commands which doing the hardware part.

````python
pv.hardware.deploy()
````
