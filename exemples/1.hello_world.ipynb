Welcom to Pyvado Hellow World Script

In this script, we will see basic command of pyvado.

First, import pyvado

```python
from pyvado import Pyvado
```

Now we will open our pyvado session

```python
pv = Pyvado()
```

Perfect! We will run our first TCL command. A simple puts command.

```python
pv.tcl.run(
  cmd = "puts \"Hello Pyvado!\""
)
```

Now we will read the output of vivado consol

```python
pv.tcl.read()
```

Congrats! You run your first command with Pyvado!

Now we will run two commands. To do that we can call run_command two times or we can send a list of commands

```python
cmd = [
  "puts \"Command 1\"",
  "puts \"Command 2\""
]
pv.tcl.run(
  cmd = cmd
)
```

```python
print(f"result of command 1 {pv.tcl.read()}")
print(f"result of command 2 {pv.tcl.read()}")
```

Perfect!
Sometimes, vivado commands take some times to run. Let's try to run a long command. 
We will first wait for 5 seconds and then send a message. Normaly the message should be read after 5 seconds. Let's try it!

```python
import time 

start = time.time()
pv.tcl.run("after 5000")
end = time.time()
pv.tcl.run("puts \"message\"")
print(f"message read after {end-start} seconds : {pv.tcl.read()}")
```
